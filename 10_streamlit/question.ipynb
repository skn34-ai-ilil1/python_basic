문제 2. 사용자 정보 입력 폼 만들기 (question/1_input_page.py)
다음 조건을 만족하는 Streamlit 페이지 pages/1_input_page.py를 작성하세요.
페이지 제목을 사용자 입력 받는 페이지로 표시하세요.
st.columns를 사용해 3개의 컬럼을 만들고, 왼쪽 첫 번째 컬럼(col1)에 다음 입력 요소들을 배치하세요.
닉네임을 입력받는 st.text_input (레이블: 닉네임 입력)
나이를 입력받는 st.number_input (레이블: 나이 입력, 최소값 13, 최대값 100)
국적을 선택하는 st.selectbox
옵션: ['한국', '중국', '일본', '미국', '외계']
취미를 선택하는 st.radio
옵션: ['맛집 탐방', '영화 감상', '음악 감상', '뜨개질']
개인정보 제공 동의 여부를 체크하는 st.checkbox (레이블: 개인정보 제공 동의)
오른쪽 컬럼(col2)에는 " 입력 완료" 라는 버튼을 하나 만들고, 버튼이 눌리면 다음 내용을 st.write()로 출력하세요.
이름이 뭐야? {닉네임}
몇 살이야? {나이}
어디서 왔어? {국적}
취미가 뭐야? {취미}
개인정보 제공에 동의해? {체크 여부}
