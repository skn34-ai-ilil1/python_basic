문제 2: 카피바라의 업무 관리 패키지
카피바라는 팀의 업무를 관리하기 위한 패키지를 작성하려고 합니다.
work_management라는 패키지를 생성하고, 아래와 같은 모듈을 포함하세요:
task_tracking.py: 작업 상태를 관리하는 모듈
함수 start_task(task):"작업 '{task}' 시작됨." 반환
함수 end_task(task): "작업 '{task}' 종료됨." 반환
​
reporting.py: 업무 보고를 관리하는 모듈
함수 generate_report(task): "'{task}' 작업 보고서 생성됨." 반환
​
메인 스크립트에서 이 패키지를 불러와 아래와 같이 출력하세요:
작업 '코드 리뷰' 시작됨.
작업 '코드 리뷰' 종료됨.
'코드 리뷰' 작업 보고서 생성됨.


```python
import work_management.task_tracking
import work_management.reporting

print(work_management.task_tracking.start_task("데이터 분석"))
print(work_management.task_tracking.end_task("데이터 분석"))
print(work_management.reporting.generate_report("데이터 분석"))
```

문제 3: 카피바라의 개인 일정 관리 패키지
카피바라는 자신의 일정을 관리하기 위한 패키지를 작성하려고 합니다.
schedule_management라는 패키지를 생성하고, 아래와 같은 모듈을 포함하세요:
calendar.py: 카피바라의 일정을 관리하는 모듈
함수 add_event(event, date):"{date}에 '{event}' 일정이 추가되었습니다." 반환
함수 remove_event(event):"'{event}' 일정이 삭제되었습니다." 반환
​
메인 스크립트에서 이 패키지를 불러와 아래와 같이 출력하세요:
2024-01-15에 '회의' 일정이 추가되었습니다.
'회의' 일정이 삭제되었습니다.

```python
import schedule_management.calendar
import datetime

print(schedule_management.calendar.add_event("회의", datetime.date(2024, 1, 15)))
print(schedule_management.calendar.remove_event("회의"))
```
