문제 1: 손님 응대 중단하기
"카피바라는 갑자기 피곤해져서 손님 응대를 중단하고 싶다."
카피바라는 카페에서 손님을 맞이하고 있습니다. 하지만 카피바라는 너무 피곤해서 3번째 손님까지만 응대하고 싶습니다.
손님 리스트 customers에 '토끼', '곰', '여우', '호랑이', '늑대'를 저장하세요.
for문을 사용해 각 손님을 응대하며 다음 메시지를 출력하세요:
"안녕하세요, [손님 이름]님! 무엇을 도와드릴까요?"
3번째 손님까지 응대한 후, 반복문을 종료하고 "죄송합니다, 오늘은 여기까지입니다!"를 출력하세요.


```python
customers = ["토끼", "곰", "여우", "호랑이", "늑대"]

count = 0

for customer in customers:
    if count < 3:
        print(f"안녕하세요, {customer}님! 무엇을 도와드릴까요?")
        count += 1
        print(count)
    else:
        print("죄송합니다, 오늘은 여기까지입니다!")
        break
```


문제 2: 주문 실수 중단하기
"카피바라는 주문을 받다가 실수를 발견하면 즉시 중단하려고 한다."
카피바라는 손님들의 주문을 확인하고 있습니다. 주문에 문제가 있으면 즉시 중단해야 합니다.
주문 리스트 orders에 '아메리카노', '카페라떼', '핫초코', '모르겠어요', '에스프레소'를 저장하세요.
for문을 사용해 각 주문을 확인하며 다음과 같이 출력하세요.
"[음료 이름] 주문 확인했습니다!"
만약 주문이 '모르겠어요'이면, "주문에 문제가 있습니다! 확인이 필요합니다."를 출력하고 반복문을 종료하세요.


```python
orders = ['아메리카노', '카페라떼', '핫초코', '모르겠어요', '에스프레소']

for i in orders:
    if i == "모르겠어요":
        print("주문에 문제가 있습니다! 확인이 필요합니다.")
        break
    else:
        print(f"{i} 주문 확인했습니다!")
```


문제 3: 홀수 테이블 청소 건너뛰기
"카피바라는 짝수 테이블만 청소하고 싶다."
카피바라는 홀수 테이블은 건너뛰고 짝수 테이블만 청소하려고 합니다.
카피바라는 홀수 테이블은 건너뛰고 짝수 테이블만 청소하려고 합니다.
테이블 리스트는 1부터 6까지의 숫자로 구성되어 있습니다.
for문을 사용해 각 테이블 번호를 확인하세요.
테이블 번호가 홀수이면 해당 테이블을 건너뛰세요.
짝수 테이블인 경우 "[n번 테이블] 청소 완료!"를 출력하세요.

```python
table_list = [i for i in range(1,7)]

print(table_list)

for i in table_list:
    if i % 2 != 0:
        continue
    else :
        print(f"[{i}번테이블] 청소 완료!")
```
