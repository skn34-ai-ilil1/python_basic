
문제 8: 주말 계획
"카피바라의 주말 계획은?"
주말이 되자 카피바라는 여유로운 시간을 보내려 합니다.
사용자에게 "오늘 날씨가 어떤가요? (맑음/비/눈): "를 입력받으세요.
입력값이 '맑음'이면 "한강에 가서 피크닉을 즐기는 척하며 일을 한다."를 출력하세요.
입력값이 '비'이면 "집에서 빗소리를 ASMR 삼아 일을 한다."를 출력하세요.

```python
print("====================================")
print("오늘 날씨가 어떤가요? (맑음/비/눈): ")
str4 = input()
print("====================================")

if str4 == "맑음":
    print("한강에 가서 피크닉을 즐기는 척하며 일을 한다.")
elif str4 == '비' :
    print("집에서 빗소리를 ASMR 삼아 일을 한다.")
elif str4 == '눈' :
    print("전기장판 안에서 귤을 까먹으며 일을 한다.")
else :
    print("암막 커튼 때문에 일어나지 못했다. 푹 잘 수 있군, 럭키비키잖아?")
```
