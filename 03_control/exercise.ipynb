```python
print("=== OhGiraffers vending machine ===")
print("tiger: 우리는 1인 24시간 교대제로 간다")
print("monkey: 24시간 문의 환영, 다른 말로 하면?")
print("squirrel: 저도 하루에 너다섯 시간 자는데요?")
print("capybara: 살아남는 사람이 강한 사람입니다.")
print("====================================")
print("만나고 싶은 선생님을 입력해 주세요")
input_teacher = input()
print("====================================")

sleep_time = 0
match input_teacher:
    case "tiger":
        print("호랑이 선생님의 메타버스에 오신 것을 환영합니다.")
        sleep_time = 3
    case "monkey":
        print("원숭이 선생님의 밀림에 오신 것을 환영합니다.")
        sleep_time = 4
    case "squirrel":
        print("다람쥐 선생님의 매력의 늪에 오신 것을 환영합니다.")
        sleep_time = 5
    case "capybara" :  
        print("카피바라 선생님의 교실에 오신 것을 환영합니다.")
        sleep_time = 6
    case _:
        print("공부할 마음이 없으신 건 아니죠?")

print("매일 " + str(sleep_time) + "시간의 수면이 예상됩니다.")
# print("매일", sleep_time, "시간의 수면이 예상됩니다.")
```

```python
a = int(input('첫번째 정수 입력 : '))
b = int(input('두번째 정수 입력 : '))

big = a if a > b else b
small = a if a < b else b
print(big, small)
```

```python
fruits = ['apple', 'kiwi', 'banana', 'pear', 'apple', 'orange']
fruits.reverse()
print(fruits)   
```

```python
abclist = ['A','B','C','D','E','F']


del abclist[1:3]  # 1번째 인덱스에서 3번째까지 제거
print(abclist)    # ['A', 'D', 'E', 'F']

del abclist[:]    # 모든 요소 제거
print(abclist)    # []
```

```python
num = 0
i = 1

while i < 8:
    if i % 3 == 0:
        break
    i += 1
    num += i

print(num)
```
