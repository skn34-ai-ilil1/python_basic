문제 1: 손님 응대하기
"카피바라는 카페에서 아르바이트를 시작했다."
카피바라는 손님들에게 인사를 해야 합니다.
손님 리스트 customers에 다음과 같은 손님들이 있습니다.
'토끼', '곰', '여우', '호랑이'
for문을 사용하여 각 손님에게 다음과 같은 메시지를 출력하세요:
"안녕하세요, [손님 이름]님! 무엇을 도와드릴까요?"

```python
customers = ['토끼', '곰', '여우', '호랑이']
for i in customers:
    print(f"안녕하세요 {i}님 무엇을 도와드릴까요?" )
```


문제 2: 메뉴 준비하기


문제 7: 테이블 청소하기
"카피바라는 닫기 전에 테이블을 모두 청소해야 한다."
카피바라는 5번 테이블까지 순서대로 청소해야 합니다.
while문을 사용하여 1번부터 5번 테이블을 청소하세요.
각 테이블마다 "[n번 테이블] 청소 완료!"를 출력하세요.


```python
i = 1
while i <= 5:
    print(f"{i}번 테이블 청소완료")
    i+=1
```


문제 8: 손님이 그만 올 때까지 기다리기
"손님이 그만 올 때까지 카운터에서 기다리는 카피바라."
카피바라는 손님이 그만 올 때까지 카운터를 지켜야 합니다.
변수 customers에 초기값 5를 설정하세요.
while문을 사용해 손님이 줄어들 때마다 customers를 1씩 감소시키세요.
손님이 0이 될 때까지 "남은 손님 수: [customers]명"을 출력하세요.
손님이 모두 떠나면 "오늘의 영업이 끝났습니다!"를 출력하세요.


```python
customers = 5
while customers >= 0:

    if(customers == 0):
        print("오늘의 영업이 끝났습니다")
        break
    
    else:
        print(f"남은 손님수 [{customers}]명")
        customers-=1
```


문제 9: 알바비 계산하기
"카피바라는 일한 시간만큼 알바비를 계산합니다."
카피바라의 시급은 1시간에 1000원입니다.
변수 work_hours에 일한 시간(정수)을 입력받으세요.
while문을 사용해 일한 시간만큼 반복하며 pay에 시급을 더하세요.
최종 알바비를 "오늘의 알바비는 [pay]원입니다."로 출력하세요.

```python
work_hours = int(input())
pay = 0
while work_hours > 0:
    work_hours = work_hours -1
    print(work_hours)
    pay = pay+1000

print(f"오늘의 알바비는 {pay}원 입니다")

```
