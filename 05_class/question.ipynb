문제 1: 클래스 속성, 인스턴스 속성 테스트
카피바라는 이직 준비를 위해 자기소개서를 작성해야 합니다. Squirrel이라는 클래스를 만들어 카피바라의 기본 정보를 설정하세요.
클래스 속성 species를 설정하여 "카피바라"로 초기화하세요.
생성자를 통해 인스턴스 속성 name과 current_job을 설정하세요.
메서드 introduce를 만들어 이름과 현재 직업을 소개하는 문자열을 반환하세요.
다른 인스턴스를 생성해도 species는 공유되지만, name과 current_job은 서로 독립적임을 확인하세요.

```python
class Squirrel:
    species = '카피바라'

    def __init__(self, name, current_job):
        self.name = name
        self.current_job = current_job

    def introduce(self):
        return f'저는 {self.name}입니다. 현재 직업은 {self.current_job} 종 {self.species}입니다.'


sq = Squirrel('초코', '학생')
sq2 = Squirrel('초코2', '학생2')

print(sq.introduce())
print(sq2.introduce())
```


문제 2: 클래스 속성과 인스턴스 속성의 차이
카피바라는 이직 준비를 하며 다양한 직업 스킬을 학습 중입니다. 클래스 속성과 인스턴스 속성의 차이를 확인하세요.
Character 클래스를 생성하세요.
클래스 속성 skills를 정의하여 빈 리스트로 초기화하세요.
생성자에서 인스턴스 속성 skills를 별도로 초기화하세요.
add_skill 메서드에서 클래스 속성이 아닌 인스턴스 속성에 스킬을 추가하도록 수정하세요.

```python
class Character:
    skills = []
    def __init__(self, name, current_job):
        self.name = name
        self.current_job = current_job

    def add_skill(self, skill):
        self.skills.append(skill)

    def show_skill(self):
        return f'{self.name}의 스킬: {", ".join(self.skills)}'
    
class Character2:
    
    def __init__(self, name, current_job, skills=None):
        self.name = name
        self.current_job = current_job
        self.skills = skills if skills is not None else []

    def add_skill(self, skill):
        self.skills.append(skill)

    def show_skill(self):
        return f'{self.name}의 스킬: {", ".join(self.skills)}'
    
my_character = Character('산골짜기다람쥐', '전사')
your_character = Character('도시카피바라', '법사')
my_character.add_skill('곤봉 휘두르기')
your_character.add_skill('마법진 그려서 날려버리기')

print(my_character.show_skill())
print(your_character.show_skill())


my_character = Character2('산골짜기다람쥐', '전사')
your_character = Character2('도시카피바라', '법사')
my_character.add_skill('곤봉 휘두르기')
your_character.add_skill('마법진 그려서 날려버리기')

print(my_character.show_skill())
print(your_character.show_skill())
```


문제 4: 네임스페이스 테스트 - 전역과 로컬
카피바라는 이직 준비 도구를 사용하며, 전역 변수와 로컬 변수를 관리합니다. namespace_test 함수를 작성해 네임스페이스를 테스트하세요.
전역 변수 tool을 "연필"로 설정하세요.
 함수 namespace_test 내부에 로컬 변수 local_tool을 "키보드"로 설정하고, global 키워드를 사용해 전역 변수를 "펜"으로 수정하세요.
함수 내부에서 로컬 변수와 전역 변수의 값을 출력하세요.

```python
tool = "연필"

def namespace_test():
    local_tool ="키보드"
    print(local_tool)
    global tool
    tool = "펜"

print(tool)    
namespace_test()
print(tool)   
```


문제 5: 네임스페이스 테스트 - 클래스와 인스턴스
카피바라는 이직 준비 중 클래스를 통해 도구를 관리합니다. Toolbox 클래스를 만들어 네임스페이스를 테스트하세요.
클래스 속성 tools를 ["노트북", "책상"]으로 초기화하세요.
생성자를 통해 인스턴스 속성 tools를 설정하고, 객체마다 다른 도구를 가질 수 있도록 설정하세요.
add_tool 메서드를 만들어 클래스 속성에 새 도구를 추가하세요.

```python
class Toolbox:
    def __init__(self, tools=None):
        if tools is None:
            self.tools = ['노트북','책상']
        else:
            self.tools = tools

    def add_tool(self, tool):
        self.tools.append(tool)

    def show_tools(self):
        return f'도구함에 있는 도구: {", ".join(self.tools)}'
    
toolbox1 = Toolbox()
toolbox2 = Toolbox(['마우스','키보드'])

print(toolbox1.show_tools())
print(toolbox2.show_tools())

toolbox1.add_tool('펜')
print(toolbox1.show_tools())
```


문제 6: 네임스페이스 테스트 - nonlocal
카피바라는 이직 준비 중 추가적인 정보가 필요합니다. 중첩 함수와 nonlocal 키워드를 사용해 바깥 함수의 변수를 수정하세요.
함수 nested_function을 작성하세요.
바깥 함수의 변수 status를 "작성 중"으로 초기화하세요.
중첩 함수에서 status를 "완료"로 수정하세요.

```python
def nested_function():
    status = "작성 중"

    def inner_function():
        nonlocal status
        status = "완료"

    print(status)
    inner_function()
    print(status)

nested_function()
```
