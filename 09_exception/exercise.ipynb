```python
f = None

try:
    f = open('foo.txt', 'w', encoding="utf-8")
    f.write("멍청이")
finally:
    if f:
        f.close()   # try 블럭 수행 중간에 오류가 발생하더라도 마지막에 무조건 실행된다.
```

```python
try:
    age=int(input('나이를 입력하세요: '))
except:
    print('입력이 정확하지 않습니다.')
else:
    if age <= 18:
        print('미성년자는 출입금지입니다.')
    else:
        print('환영합니다.')
```
