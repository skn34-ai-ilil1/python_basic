
문제 2: 특정 예외 처리
카피바라는 사용자에게 나이를 입력받아 성인인지 확인하려고 합니다.
사용자로부터 나이를 입력받아 정수로 변환하세요.
입력값이 숫자가 아닐 경우, "숫자를 입력해야 합니다."를 출력하세요.
숫자가 올바르게 입력되면 나이가 19 이상인지 확인하고, 19 이상이면 "성인입니다.", 그렇지 않으면 "미성년자입니다."를 출력하세요.

```python

try:
    age=int(input('나이를 입력하세요: '))
except:
    print('숫자를 입력해야 합니다.')
else:
    if age <= 18:
        print('미성년자입니다.')
    else:
        print('성인입니다.')
```


문제 3: 특수 예외 상황과 오류 변수 사용
카피바라는 숫자 두 개를 입력받아 나누기를 수행하려고 합니다.
사용자로부터 두 숫자를 입력받아 나누기를 수행하세요.
첫 번째 숫자가 0으로 나누기를 시도하면 "0으로 나눌 수 없습니다."를 출력하세요.
입력값이 숫자가 아니면 "숫자를 입력해야 합니다."를 출력하세요.
오류 메시지를 출력할 때 발생한 오류 내용을 함께 출력하세요.


```python
try:
    n1 = int(input())
    n2 = int(input())
    n1/n2
except ZeroDivisionError as e:
    print("0으로 나눌 수 없습니다.")
    print(e)
except:
    print("숫자를 입력해야 합니다.") 
```


문제 4: finally 블럭
카피바라는 프로그램 실행 후 파일 자원을 반드시 정리해야 합니다.
log.txt 파일을 읽으려고 시도하세요.
파일을 여는 중 예외가 발생할 경우, "파일을 열 수 없습니다."를 출력하세요.
예외 발생 여부와 상관없이 파일 자원을 정리하는 메시지 "파일 자원 정리 완료"를 출력하세요.


```python
try:
    with open("log.text", "r", encoding='utf-8') as f:
        contents = f.read()
        print(contents)
except:
    print('파일을 열 수 없습니다')
finally:
    print("파일 자원 정리 완료")      
```


문제 5: try-else 블럭
카피바라는 사용자로부터 숫자를 입력받아 2로 나눈 결과를 출력하려고 합니다.
사용자로부터 숫자를 입력받아 정수로 변환하세요.
입력값이 숫자가 아닐 경우 "숫자를 입력해야 합니다."를 출력하세요.
숫자가 올바르게 입력되면 입력값을 2로 나눈 결과를 출력하세요.

```python
try:
    n1 = int(input())
except:
    print("숫자를 입력해야 합니다.")  
else:
    print(n1/2)
```

```python
def solution(storage, usage, change):
    total_usage = 0
    for i in range(len(change)):
        usage = total_usage * change[i]/100
        total_usage += usage
        if total_usage > storage:
            return i
    
    return -1
```

```python
from enum import Enum
# name = value 지정 필수
class Gender(Enum):
    M = '남'
    F = '여'

print(Gender.M)
print(Gender.F)

# name
print(Gender.M.name)
print(Gender.F.name)

# value
print(Gender.M.value)
print(Gender.F.value)

for gender in Gender:
    print(gender)


Color = Enum('Color', ['RED', 'GREEN', 'BLUE'])
for color in Color:
    print(color.name, color.value)

from enum import Enum, auto

class Lang(Enum):
    C = auto()        # 1
    Java = auto()     # 2
    Python = auto()   # 3

for lang in Lang:
    print(lang.name, lang.value)
```

```python
from enum import Enum

class Number(Enum):
    ONE = 1
    FIRST = 1   # ONE의 별칭(alias)

print(Number.ONE)    # Number.ONE
print(Number.FIRST)  # Number.ONE
print(Number.ONE == Number.FIRST)   # True
print(Number.FIRST is Number.ONE)   # True

for num in Number:
    print(num)   # 대표 멤버만 순회
```
